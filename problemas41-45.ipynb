```python
from math import floor, sqrt

def issquare(c):
    if floor(sqrt(c)) == sqrt(c):
        return(True)
    else:
        return(False)
    






f = open("problema42.txt", "r")
array = []
line = f.readline()

array = [x for x in line.split('","')]
f.close()


def triangulo(n):
    return(issquare(1+8*n))
    #un numero es triangular sii 1+8*n es cuadrado.
    
    
def valor(a):
    suma =0
    for i in range(len(a)):
        #algunos caracteres tienen " y no cuenta
        if a[i] != '"':
            suma = suma + ord(a[i]) - 64
    return(suma)
            
tot=0
for a in array:
    if triangulo(valor(a)) == True:
        tot = tot+1
        
print(tot)
```

```python
from bisect import bisect_left

pentagonal = []
for n in range(1, 10000):
    p = int((n*(3*n - 1))/2)
    pentagonal.append(p)


def bisect_search(lst, item):
    ''' it is used for sorted lists'''
    return (item <= lst[-1]) and (lst[bisect_left(lst, item)] == item)


for i in pentagonal:
    for j in pentagonal:
        diff = abs(i - j)
        sam = i + j
        if bisect_search(pentagonal, sam) and bisect_search(pentagonal, diff):
            print("(", i, ":", j, ")")
```

```python
import random

def funcc():
    nums=[]
    for i in range(8):
        nums.append(random.randint(1,10))
    numset = set(nums)
    print(numset, len(numset))
    return(len(numset) == 8)
```

```python
suma = 0
for i in range(10000):
    if funcc()==True:
        suma += 1
print(suma/10000)
```

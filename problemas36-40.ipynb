## 36. Palindromos de Doble Base
Encuentra todos los numeros menores a 1,000,000 que son palindromos en base 10 y base 2

```python
bin(1324)

def abin(n):
    s = str(bin(n))
    return(int(s[2:]))

def checa(n):
    s = str(n)
    a = s[::-1]
    if a == s:
        return(True)
    else:
        return(False)


res=[]
for i in range(1000000):
    if checa(i)==False:
        continue
    if checa(abin(i)) == False:
        continue
    res.append(i)
print(res)
print(sum(res))
```

## 37. Primos Truncables

Hay 11 primos que son truncables a la izquierda y a la derecha. Por ejemplo 3797 es primo pero tambien 379, 37, 3, por lo que es truncable derecha. Y tambien 797,97,7 son primos. <br>
Buscamos los 11 primos con estas caracter'isticas

```python
import sympy

def trunci(n):
    n = str(n)
    for i in range(0,len(n)):
        if sympy.isprime(n[i:]) == False:
            return(False)
    return(True)

def truncd(n):
    n = str(n)
    for i in range(1,len(n)):
        if sympy.isprime(n[:i]) == False:
            return(False)
    return(True)



res =[]

i=23

while len(res) < 11:
    if trunci(i) == False:
        i=i+1
        continue
    if truncd(i)== False:
        i=i+1
        continue
    res.append(i)
    print(i)
    i=i+1
print(res)
print(sum(res))
    
    
```

# 38. Multiplos Pandigitales


```python
# Primero, el numero tiene que ser menor a 9999

def checadifdig(n):
    #checa si un numero n no tiene repeticiones
    ns = str(n)
    ns = [int(x) for x in ns]
    nss = set(ns)
    #para lo cual, su str tiene que ser igual a su set
    if len(nss)  == len(ns):
        return(True)
    else:
        return(False)

    
res=[]
#va iterando
for k in range(9999):
    #para empezar, el numero tiene que tener digitos distintos
    if checadifdig(k) == True:
        conc = ''
        i=1
        #en ese caso, vamos a ir concatenando los productos de k
        #siempre y cuando tenga menos de 9 elementos la concatenacion
        while len(conc) < 9:
            conc = conc + str(i*k)
            i=i+1
    #luego, si esta concatenacion tiene exactamente 9 elementos, nos fijamos 
        if len(conc) == 9:
            #vemos que sean distintos.
            if checadifdig(int(conc)) == True and ('0' in conc)==False:
                res.append([k,conc])

print(checadifdig(9353))            
print(res)
    
ress=[]
for i in range(len(res)):
    ress.append(res[i][-1])
print("\n")
print(ress)
print(max(ress))
```

```python
from math import sqrt, floor


def issquare(c):
    if floor(sqrt(c)) == sqrt(c):
        return(True)
    else:
        return(False)
    


def solve():
    """ Compute the answer to Project Euler's problem #39 """
    upper_limit = 1000
    #answers guarda la cantidad de soluciones con cierto perimetro
    answers = [0] * (upper_limit + 1)
    #iteramos sobre todas las a
    for a in range(1, upper_limit + 1):
        # y sobre todas las b > a
        for b in range(a, upper_limit + 1):
            # c es la hipotenusa
            c2 = a * a + b * b
            #c2 tiene que ser tal que a+b+c < perimetro, sino no nos interesa, y tiene que ser cuadrado
            if c2 <= (upper_limit - a - b) * (upper_limit - a - b) and issquare(c2):
                # a + b + c <= upper_limit
                # c <= upper_limit - a - b
                # c^2 <= (upper_limit - a - b)^2
                
                #agragamos un 1 a nuestra respuesta.
                c = int(sqrt(c2))
                p = a + b + c
                if p <= upper_limit:
                    answers[p] += 1
    answer = max(range(len(answers)), key=lambda _p: answers[_p])  # find arg-max(answers)
    return answer
print(solve())
    
```
